```python
import numpy as np
import pandas as pd
```

```python
# read the csv file
df_features = pd.read_csv('./features_2013-2020.csv')
df_features.columns
```

## Calaulate history win probability

```python
# order the row by date, because the race id is different when it comes to the date
# Convert date column to datetime format
df_features['date'] = pd.to_datetime(df_features['date'])

# Sort DataFrame by date
df_features = df_features.sort_values(by='date',ascending=True)
df_features
```

```python
# distribute the phi id in order
df_features = df_features.sort_values(by=['date', 'race_id']).reset_index(drop=True)

df_features['race_id'] = df_features.groupby(['date', 'race_id']).ngroup() + 1
```

```python
df_features['horse_win_rate'] = 0.0
df_features['jockey_win_rate'] = 0.0
for idx, row in df_features.iterrows():
    horse_id = row['horse_id']
    previous_races = df_features[(df_features['horse_id'] == horse_id) & (df_features.index < idx)]
    if len(previous_races) > 0:
        win_rate = previous_races['won'].sum() / len(previous_races)
    else:
        win_rate = 0.0
    df_features['horse_win_rate'].iloc[idx]= win_rate
    
    jockey_id = row['jockey_id']
    previous_races = df_features[(df_features['jockey_id'] == jockey_id) & (df_features.index < idx)]
    if len(previous_races) > 0:
        win_rate = previous_races['won'].sum() / len(previous_races)
    else:
        win_rate = 0.0
    df_features['jockey_win_rate'].iloc[idx]= win_rate
    
```

```python
# horse_win_times_career
df_features['horse_win_times_career'] = 0.0
df_features['jockey_win_times_career'] = 0.0
for idx, row in df_features.iterrows():
    horse_id = row['horse_id']
    previous_races = df_features[(df_features['horse_id'] == horse_id) & (df_features.index < idx)]
    if len(previous_races) > 0:
        win_rate = previous_races['won'].sum()
    else:
        win_rate = 0.0
    df_features['horse_win_times_career'].iloc[idx]= win_rate
    
    jockey_id = row['jockey_id']
    previous_races = df_features[(df_features['jockey_id'] == jockey_id) & (df_features.index < idx)]
    if len(previous_races) > 0:
        win_rate = previous_races['won'].sum()
    else:
        win_rate = 0.0
    df_features['jockey_win_times_career'].iloc[idx] = win_rate
```

## Independent Component Analysis

```python
from sklearn.decomposition import FastICA
from sklearn.preprocessing import StandardScaler
```

```python
df_features = df_features.loc[:, ~df_features.columns.str.contains('ICA')]
```

```python
df_features_transform = df_features
# df_features_transform[features_to_transform]
```

```python
# numeric_features = ['declared_weight','actual_weight','distance','horse_win_times_career','jockey_win_times_career','finish_time_seconds']
numeric_features = ['declared_weight','actual_weight','horse_win_times_career','jockey_win_times_career']
# categorical_features = ['track_condition',"horse_id","jockey_id","trainer_id","location","horse_class"]
categorical_features = ["horse_id","jockey_id","trainer_id"]

# one hot encode
df_features_transform = pd.get_dummies(df_features_transform, columns=categorical_features, drop_first=True)

# Standardize the numeric features
scaler = StandardScaler()
df_features_transform[numeric_features] = scaler.fit_transform(df_features_transform[numeric_features])

```

```python
# features_to_transform = numeric_features + [col for col in df_features_transform.columns if any(prefix in col for prefix in ['track_condition_', 'horse_id_', 'jockey_id_', 'location_', 'horse_class_'])]
features_to_transform = numeric_features + [col for col in df_features_transform.columns if any(prefix in col for prefix in ['trainer_id', 'horse_id_', 'jockey_id_'])]
```

```python
n_component = 3
ica = FastICA(n_components=n_component, random_state=42)
ica_features = ica.fit_transform(df_features_transform[features_to_transform])
```

```python
ica_features_df = pd.DataFrame(ica_features, columns=[f'ICA_{i+1}' for i in range(n_component)])
df = pd.concat([df_features, ica_features_df], axis=1)
```

```python
df_features = df
```
